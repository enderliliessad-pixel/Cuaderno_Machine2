```python
import os
import numpy as np

ruta = "procesado_npy/train/Tomato___healthy"

# obtener un archivo real
archivo = os.listdir(ruta)[0]

print("Archivo encontrado:", archivo)

img = np.load(os.path.join(ruta, archivo))

print(img.shape)
print(img.min(), img.max())
```

```python
import matplotlib.pyplot as plt

ruta = "procesado_gray/train/Tomato___healthy"

archivo = os.listdir(ruta)[0]  # toma una imagen cualquiera
img = np.load(os.path.join(ruta, archivo))
print(img.shape)   # debe salir (224, 224, 1)
print(img.shape)
print(img.min(), img.max())

plt.imshow(img.squeeze(), cmap="gray")
plt.title(archivo)
plt.axis("off")
plt.show()
```

```python
ruta_rgb = "procesado_npy/train/Tomato___healthy"
ruta_gray = "procesado_gray/train/Tomato___healthy"

archivo_rgb = os.listdir(ruta_rgb)[0]
archivo_gray = os.listdir(ruta_gray)[0]

img_rgb = np.load(os.path.join(ruta_rgb, archivo_rgb))
img_gray = np.load(os.path.join(ruta_gray, archivo_gray))

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("RGB")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_gray.squeeze(), cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.show()
```
