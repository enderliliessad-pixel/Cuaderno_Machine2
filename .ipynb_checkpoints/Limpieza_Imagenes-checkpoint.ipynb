## Eliminaremos las imágenes que puedan estar dañadas

### Lo haremos para train y para val

```python
##Creamos las rutas 

ruta_img_train = "recursos/img/train"
ruta_img_test = "recursos/img/val"
```

```python
import os
from PIL import Image

def eliminar_imagenes_corruptas(ruta_img):
    eliminadas = 0
    
    for root, _, files in os.walk(ruta_img):
        for file in files:
            path = os.path.join(root, file)
            
            try:
                with Image.open(path) as img:
                    img.verify()  # con el verify podremos saber si está dañada
            except:
                print(f"Eliminando corrupta: {path}")
                os.remove(path)
                eliminadas += 1
    
    print(f"Total eliminadas: {eliminadas}")
```

```python
eliminar_imagenes_corruptas(ruta_img_train)
eliminar_imagenes_corruptas(ruta_img_test)
#No debería salir ninguna porque el dataset ya estaba limpio, pero me da pereza revisar 1000 imágenes por cada enfermedad xd
```

## Ahora vamos a eliminar imágenes que puedan estar duplicadas

```python
import hashlib

def hash_imagen(path):
    with open(path, 'rb') as f:
        return hashlib.md5(f.read()).hexdigest()

def eliminar_duplicadas(ruta_img):
    hashes = {}
    eliminadas = 0

    for root, _, files in os.walk(ruta_img):
        for file in files:
            path = os.path.join(root, file)
            h = hash_imagen(path)

            if h in hashes:
                print(f"Duplicada eliminada: {path}")
                os.remove(path)
                eliminadas += 1
            else:
                hashes[h] = path

    print(f"Total duplicadas eliminadas: {eliminadas}")
```

```python
eliminar_duplicadas(os.path.join(ruta_img_train))
eliminar_duplicadas(os.path.join(ruta_img_test))
```

## Filtrar imágenes para ver si están claras u oscuras

```python
import cv2
import numpy as np

def es_imagen_mala(path, umbral_min=30, umbral_max=220):
    img = cv2.imread(path)
    if img is None:
        return True
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    brillo = np.mean(gray)

    if brillo < umbral_min or brillo > umbral_max:
        return True
    
    return False

def eliminar_imagenes_malas(ruta_dataset):
    eliminadas = 0

    for root, _, files in os.walk(ruta_dataset):
        for file in files:
            path = os.path.join(root, file)

            if es_imagen_mala(path):
                print(f"Eliminando mala iluminación: {path}")
                os.remove(path)
                eliminadas += 1

    print(f"Total eliminadas por iluminación: {eliminadas}")
```

```python
eliminar_imagenes_malas(ruta_img_train)
eliminar_imagenes_malas(ruta_img_test)
```

## Ahora vamos a detectar si hay imágenes borrosas

```python
import os
import cv2

def es_borrosa(path, threshold=100):
    img = cv2.imread(path)
    
    if img is None:
        return True  # imagen inválida
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    varianza = cv2.Laplacian(gray, cv2.CV_64F).var()

    return varianza < threshold

def eliminar_borrosas(ruta_dataset, threshold=100):
    eliminadas = 0

    for root, _, files in os.walk(ruta_dataset):
        for file in files:
            path = os.path.join(root, file)

            try:
                if es_borrosa(path, threshold):
                    print(f"Eliminando borrosa: {path}")
                    os.remove(path)
                    eliminadas += 1

            except Exception as e:
                print(f"Error con {path}: {e}")

    print(f"Total borrosas eliminadas en {ruta_dataset}: {eliminadas}")
```

```python
eliminar_borrosas(ruta_img_train)
eliminar_borrosas(ruta_img_test)
```

### Segmentación

```python
import cv2
import numpy as np

def segmentar_hoja(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # rango verde
    lower = np.array([25, 40, 40])
    upper = np.array([90, 255, 255])

    mask = cv2.inRange(hsv, lower, upper)

    # limpiar ruido
    kernel = np.ones((5,5), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    resultado = cv2.bitwise_and(img, img, mask=mask)

    return resultado

#Para probar
#img = cv2.imread("recursos/train/Tomato___Bacterial_spot/ffcf0c9e-f932-4836-b160-47277fbe124c___GCREC_Bact.Sp 3776.JPG")

#seg = segmentar_hoja(img)

#cv2.imshow("Original", img)
#cv2.imshow("Segmentada", seg)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

### Viendo que funciona bien, entonces aplicamos a todo el dataset de imágenes

```python
import os
import cv2
import numpy as np

def segmentar_hoja(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    lower = np.array([25, 40, 40])
    upper = np.array([90, 255, 255])

    mask = cv2.inRange(hsv, lower, upper)

    kernel = np.ones((5,5), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    resultado = cv2.bitwise_and(img, img, mask=mask)

    return resultado


def procesar_dataset(input_dir, output_dir):
    total = 0
    guardadas = 0

    print(f"\nProcesando dataset: {input_dir} ...")

    for root, _, files in os.walk(input_dir):
        for file in files:
            
            if not file.lower().endswith(('.png', '.jpg', '.jpeg')):
                continue

            total += 1
            path = os.path.join(root, file)

            img = cv2.imread(path)
            if img is None:
                continue

            seg = segmentar_hoja(img)

            rel_path = os.path.relpath(root, input_dir)
            save_dir = os.path.join(output_dir, rel_path)
            os.makedirs(save_dir, exist_ok=True)

            save_path = os.path.join(save_dir, file)

            if cv2.imwrite(save_path, seg):
                guardadas += 1

    print(f"Total encontradas: {total}")
    print(f"Total guardadas: {guardadas}")
    print("Finalizado\n")
```

```python
procesar_dataset("recursos/train", "dataset/img/img_dataset_segmentado/train")
procesar_dataset("recursos/val", "dataset/img/img_dataset_segmentado/val")
```

## Ahora vamos con los decriptores

```python
import cv2
import numpy as np
from skimage.feature import graycomatrix, graycoprops

#Descriptor de Textura
def extraer_textura(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    glcm = graycomatrix(gray,
                        distances=[1],
                        angles=[0],
                        levels=256,
                        symmetric=True,
                        normed=True)

    contraste = graycoprops(glcm, 'contrast')[0, 0]
    homogeneidad = graycoprops(glcm, 'homogeneity')[0, 0]
    energia = graycoprops(glcm, 'energy')[0, 0]
    correlacion = graycoprops(glcm, 'correlation')[0, 0]

    return [contraste, homogeneidad, energia, correlacion]


#Descriptor de forma
def extraer_forma(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    _, thresh = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)

    contornos, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contornos) == 0:
        return [0, 0]

    c = max(contornos, key=cv2.contourArea)

    area = cv2.contourArea(c)
    perimetro = cv2.arcLength(c, True)

    return [area, perimetro]
```

```python
import os
import pandas as pd

def crear_dataset_features(ruta_dataset):
    data = []

    print(f"\nExtrayendo features de: {ruta_dataset} ...")

    for clase in os.listdir(ruta_dataset):
        ruta_clase = os.path.join(ruta_dataset, clase)

        if not os.path.isdir(ruta_clase):
            continue

        for file in os.listdir(ruta_clase):
            path = os.path.join(ruta_clase, file)

            if not file.lower().endswith(('.png', '.jpg', '.jpeg')):
                continue

            img = cv2.imread(path)

            if img is None:
                continue

            textura = extraer_textura(img)
            forma = extraer_forma(img)

            features = textura + forma + [clase]

            data.append(features)

    columnas = [
        'contraste', 'homogeneidad', 'energia', 'correlacion',
        'area', 'perimetro',
        'clase'
    ]

    df = pd.DataFrame(data, columns=columnas)

    print("Extracción finalizada\n")

    return df
```

```python
train_df = crear_dataset_features("dataset/img/img_dataset_segmentado/train")
val_df = crear_dataset_features("dataset/img/img_dataset_segmentado/val")

train_df.to_csv("dataset/img/img_train_features.csv", index=False)
val_df.to_csv("dataset/img/img_val_features.csv", index=False)
```

## Guardamos el dataset de imágenes

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os

#1) Separamos las variables de train y test
X_train = train_df.drop("clase", axis=1)
y_train = train_df["clase"]

X_val = val_df.drop("clase", axis=1)
y_val = val_df["clase"]

#2) Estandarizamos tanto train como val
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)       

#3) Generamos el dataframe
columnas = train_df.drop("clase", axis=1).columns

X_train_df = pd.DataFrame(X_train, columns=columnas)
X_val_df = pd.DataFrame(X_val, columns=columnas)

#4) Agregamos la variable objetivo
X_train_df["disease_future"] = y_train.values
X_val_df["disease_future"] = y_val.values

#5) Transformamos la variable objetivo a números
disease_mapping = {
    "Tomato___healthy": 1,
    "Tomato___Bacterial_spot": 2,
    "Tomato___Early_blight": 3,
    "Tomato___Late_blight": 4,
    "Tomato___Leaf_Mold": 5,
    "Tomato___Septoria_leaf_spot": 6,
    "Tomato___Spider_mites Two-spotted_spider_mite": 7,
    "Tomato___Target_Spot": 8,
    "Tomato___Tomato_Yellow_Leaf_Curl_Virus": 9,
    "Tomato___Tomato_mosaic_virus": 10
}

X_train_df["disease_future"] = X_train_df["disease_future"].map(disease_mapping)
X_val_df["disease_future"] = X_val_df["disease_future"].map(disease_mapping)

#6) Verificamos que todo esté bien
print("Train nulos:", X_train_df["disease_future"].isnull().sum())
print("Val nulos:", X_val_df["disease_future"].isnull().sum())

print("\nClases en train:", sorted(X_train_df["disease_future"].dropna().unique()))
print("Clases en val:", sorted(X_val_df["disease_future"].dropna().unique()))

print("\nEstadísticas (ver estandarización):")
print(X_train_df.describe())

#7) Generamos el dataset, en caso que no exista la ruta, ahí mismo se crea
ruta = "dataset/img/datos_limpios"
os.makedirs(ruta, exist_ok=True)

#8) Guardamos el dataset
X_train_df.to_csv(f"{ruta}/imagenes_train_final.csv", index=False)
X_val_df.to_csv(f"{ruta}/imagenes_val_final.csv", index=False)

print("\nDataset TRAIN y VAL listos")
```
