# ¿Que son las redes neuronales convolucionales (CNN)?

Las redes neuronales convolucionales son un tipo de red que se encarga principalmente de analizar datos de manera visual, trata de encontrar diferentes tipos de patrones de los datos visuales, aunque no es lo único que hace y lo que lo vuelve ventajoso sobre otros modelos es que es capaz de reconocer patrones en cualquier dato con una estructura similar a una cuadrícula, desde espectrogramas de audio hasta escáneres médicos en 3D y otros como diversos sistemas de procesamiento de imágenes (reconocimiento de imágenes, clasificación de imágenes, etiquetado de vídeo, análisis de texto) y de voz (reconocimiento de voz, procesamiento del lenguaje natural, clasificación de texto), junto con sistemas de inteligencia artificial de última generación como robots, asistentes virtuales y demás elementos.

# ¿Cómo funcionan las redes neuronales convolucionales (CNN)?

Las redes neuronales tratan de seguir el mismo proceso que seguimos todos los seres humanos a la hora de clasificar diferentes tipos de objetos o situaciones. Pues nosotros lo hacemos a través de diferentes características que tengan los objetos o situaciones como el: 
- El tamaño
- El color
- La forma
- El contorno
- Las sombras
En definitiva, en tratar de obtener todas las características que nos permiten distinguir entre diferentes tipos de elementos o situaciones.

Las CNN funcionan tanto para imágenes en 2D y en 3D y se dividen en 3 capas:
- capa de convolución
- capa de reducción
- capas densas
- Capa totalmente conectada

De manera resuminda las primeras 2 capas se encargan de obtener todas la características y datos de la imagen y la última es la encargada de realizar la clasificación y categorización de los datos de manera resumida y la última es la encargada de revisar los datos resumidos de la anterior capa y da el veredicto final.

### Capas de la CNN 

1. Capa convolucional: Es una operación que utiliza un filtro o kernel en la imagen para extraer diferentes tipos de características. La idea es tratar de generar un patrón que nos permita extraer características o patrones de las mismas. El filtro suele ser más pequeño que la imagen porque a la hora de que el filtro vaya desplazándose por la imagen vaya creando un mapa de características con mayor y menor peso
2. Capa de reducción: Se encarga de reducir el flujo de la información obtenida en la anterior capan, obtiene la información más relevante o las características más importantes de la imagen de la anterior capa, esta capa ayuda a reducir el costo computacional que tendrá el modelo, aunque también hay que decir que si se abusa del (max  pooling) también se podrán eliminar características importantes de la imagen. De manera simplificada esta capa nos ayuda a reducir el tamaño de la imagen, pero conservando las características más relevantes. 
3. Capas densas (fully connected): Esta capa calcula la puntuación de cada característica extraída en las anteriores capas y relaciona todos las características o datos que se hayan extraído anteriormente y se realiza un proceso de clasificación final. Cada neurona de la capa fully connected contiene una función no lineal y la elección de esta dependerá del tipo de tarea que va a realizar la CNN. Aunque en la actualidad por el gran coste computacional de las mismas se está buscando alternativas. 
4. Capa de Activación: Es fundamental indicar que existen las funciones de activación y se encuentran presentes en las capas de convolución y en las capas fully connected. Estas forman parte de cada neurona artificial. Su principal función es introducir la no linealidad.
Existen diferentes tipos de activación según el problema.
- Sigmoide
- Softmax
- ReLU(Rectified Linear Unit)
- Leaky ReLU

### Tipos de activación 

1. Sigmoide: La función sigmoidal es usada en la clasificación binaria donde se varía entre 0 y 1, 0 siendo una respuesta negativa y 1 siendo una respuesta positiva dada en términos de probabilidad. Como ejemplo tenemos que se ingresa la imagen de un gato y se obtiene un resultado de 0.9 lo que casi llega a uno, da como respuesta que el animal de la foto sea un gato.  
2. Softmax: La función softmax es utilizada para la calificación multiclase donde existe más de una salida, genera probabilidades normalizadas cuya suma es 1, permitiendo identificar la clase más probable.
3. ReLU: Es ampliamente utilizada por su eficiencia y eficacia, elimina el problema del desvanecimiento del gradiente y facilita el entrenamiento de redes profundas.
4. Leaky ReLU: Es una variante de ReLU que añade una pequeña pendiente en dicha zona negativa, que ayuda a impedir que todas las activaciones negativas lleguen a cero y permitir que la red siga aprendiendo, en menor medida, pero sin pausar el entrenamiento

## Entrenamiento de una CNN

El entrenamiento de la red consiste en ajustar los kernels en las diferentes capas de convolución y los pesos (valores de importancia) para reducir el error que s genere entre la predicción que del modelo y la respuesta que nosotros le hayamos proporcionado a través de las etiquetas, también existen 2 tipos de propagación que se utilizan para el análisis de los datos, que son la propagación hacia adelante y la propagación hacia atrás.

propagación hacia adelante: Hace alusión a procesos matemáticos en los cuales la información pase por todas las capas de la red para proporcionar una respuesta 
propagación hacia atrás: En esta se calcula el error y el cómo se ajustan los parámetros de la red

## ¿Qué tan efectivas son para la clasificación de imágenes?

Este modelo es extremadamente competente en la clasificación de imágenes, su capacidad de obtener y aprender características complejas a través de pixeles brutos la hace muy útil a la hora de reconocer ciertos elementos como, emociones, escenas y objetos. 

## CNN frente a redes neuronales tradicionales

| Función | Redes neuronales convolucionales (CNNs) | Redes neuronales tradicionales |
|---|---|---|
| **Arquitectura principal** | Está compuesta por capas convolucionales, capas de reducción, capas densas y capas de activación. | Se compone principalmente de capas totalmente conectadas (densas). |
| **Tipo de datos de entrada** | Es más adecuado para datos estructurados en forma de cuadrícula como imágenes, videos, o texto. | Es flexible y se adapta a varios tipos de datos, normalmente datos tabulares o vectores aplanados. |
| **Extracción de características** | Aprende automáticamente características como bordes, texturas, formas, colores, etc. mediante filtros. | Aprende características mediante conexiones directas, por lo que suele ser menos eficaz a la hora de aprender características espaciales. |
| **Compartir parámetros** | Sí, los pesos se comparten entre las diferentes ubicaciones de la entrada. | No, cada conexión tiene su propio peso único. |
| **Número de parámetros** | Generalmente, tienen menos parámetros debido al uso compartido y la agrupación de volumen, sobre todo en el caso de las entradas de alta dimensión, como las imágenes. | Puede tener un número muy elevado de parámetros, sobre todo en el caso de las entradas de alta dimensión. |
| **Invariancia traslacional** | Es intrínsecamente bueno para reconocer características independientemente de su posición exacta en la entrada. | Es más sensible a los cambios en las características de entrada, a menos que se entrene explícitamente con datos aumentados. |
| **Eficiencia computacional** | Más eficiente para datos espaciales o de imagen, ya que tiene menos parámetros y operaciones especializadas, aunque el coste puede aumentar gracias a las funciones de activación a usar | Puede ser costosa a nivel computacional para entradas de alta dimensión debido a las conexiones densas. |
| **Aplicaciones principales** | Clasificación de imágenes, detección de objetos, segmentación de imágenes, análisis de video, imágenes médicas y algunas tareas de procesamiento del lenguaje natural. | Clasificación o regresión de datos tabulares, reconocimiento de patrones simples, aproximación de funciones, procesamiento del lenguaje natural. |
| **Ventajas clave** | Excelente para datos visuales, aprende características jerárquicas, invariancia traslacional, parámetros reducidos, menos propensas al sobreajuste en datos de imagen. | Es flexible para varios tipos de datos, adecuada para datos tabulares no espaciales y más sencilla de entender conceptualmente para tareas básicas. |
| **Limitaciones clave** | Puede ser complejo de diseñar, normalmente requiere grandes conjuntos de datos para el entrenamiento y es menos eficaz para datos tabulares no espaciales. | No es ideal para datos espaciales de alta dimensión, ignora las relaciones espaciales y es propenso al sobreajuste con muchos parámetros en entradas complejas. |

## Ejmplo de una CNN 

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Cargar dataset MNIST
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("x_train:", x_train.shape)
print("y_train:", y_train.shape)
print("x_test:", x_test.shape)
print("y_test:", y_test.shape)

# 2. Normalizar los valores de píxeles a [0, 1]
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# 3. Agregar canal para CNN: (28, 28) -> (28, 28, 1)
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

print("x_train nuevo shape:", x_train.shape)
print("x_test nuevo shape:", x_test.shape)

# 4. Crear modelo CNN
model = models.Sequential([
    layers.Input(shape=(28, 28, 1)),

    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  # 10 clases: 0 a 9
])

# 5. Compilar modelo
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 6. Ver resumen del modelo
model.summary()

# 7. Entrenar
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.1
)

# 8. Evaluar en test
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nPrecisión en test: {acc:.4f}")
print(f"Pérdida en test: {loss:.4f}")
```

```python
# Elegir una imagen del test
indice = 5

imagen = x_test[indice]
etiqueta_real = y_test[indice]

# Predecir
predicciones = model.predict(np.expand_dims(imagen, axis=0), verbose=0)
clase_predicha = np.argmax(predicciones)
confianza = np.max(predicciones)

# Mostrar
plt.imshow(imagen.squeeze(), cmap='gray')
plt.title(f"Real: {etiqueta_real} | Predicción: {clase_predicha} | Confianza: {confianza:.4f}")
plt.axis('off')
plt.show()
```

```python
def predecir_y_mostrar(model, x_test, y_test, indice):
    imagen = x_test[indice]
    etiqueta_real = y_test[indice]

    predicciones = model.predict(np.expand_dims(imagen, axis=0), verbose=0)
    clase_predicha = np.argmax(predicciones)
    confianza = np.max(predicciones)

    print("Índice:", indice)
    print("Etiqueta real:", etiqueta_real)
    print("Predicción:", clase_predicha)
    print("Confianza:", confianza)
    print("Probabilidades:", predicciones[0])

    plt.imshow(imagen.squeeze(), cmap='gray')
    plt.title(f"Real: {etiqueta_real} | Predicción: {clase_predicha} | Confianza: {confianza:.4f}")
    plt.axis('off')
    plt.show()

# Ejemplo
predecir_y_mostrar(model, x_test, y_test, 5)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('Precisión del modelo')
plt.xlabel('Época')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from tensorflow.keras.utils import load_img, img_to_array

def cargar_imagen_externa(ruta):
    # Cargar en escala de grises y redimensionar
    img = load_img(ruta, color_mode='grayscale', target_size=(28, 28))
    img = img_to_array(img)

    # Normalizar
    img = img.astype("float32") / 255.0

    return img

# Imagen nuestra
ruta = "images/Numero0.jpeg"
img_nueva = cargar_imagen_externa(ruta)

pred = model.predict(np.expand_dims(img_nueva, axis=0), verbose=0)
clase_predicha = np.argmax(pred)
confianza = np.max(pred)

plt.imshow(img_nueva.squeeze(), cmap='gray')
plt.title(f"Predicción: {clase_predicha} | Confianza: {confianza:.4f}")
plt.axis('off')
plt.show()
```
